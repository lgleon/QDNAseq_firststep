```python
import os
import glob

def ensure_dir(file_path):
    directory = os.path.dirname(file_path)
    if not os.path.exists(directory):
        os.makedirs(directory)

def extract_filename(fname):
    fn = os.path.basename( fname )
    pre, ext = os.path.splitext(fn)
    return pre
```

```python
import subprocess

def run_ext_command(cmd):
    """This function runs an extermal command, defined by the cmd paramter
        params:
            :cmd: is the full command to execute as a string
    """
    p = subprocess.Popen(cmd, shell=True, stdout=subprocess.PIPE, stderr=subprocess.STDOUT)
#    p = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT)
    for line in p.stdout.readlines():
        print(line)

    errcode = p.wait()
    print("error code", errcode)
    return errcode

def run_bwa( sample, output, cores=T, bwaindex=BWA_INDEX):
    """ this function executes BWA-aln using the location in the filesystem of the sample and an output directory where results will be placed
    params:
        sampleA: path to sample A
        outputdir: path to the output directory
        bindex: bwa indexes needs in the alignment, default var = BWA_INDEX, defined at the begining
    """
    cmd = " bwa aln -t {0} {1} {2} > {3}".format(T, BWA_INDEX, sample, output)

    run_ext_command( cmd )
```
